```python
# you need to put your credentials into the environment variables:
# I put them into credentials.py

import os
import credentials
os.environ["AZURE_OPENAI_API_KEY"] = credentials.AZURE_OPENAI_API_KEY
os.environ["AZURE_OPENAI_ENDPOINT"] = credentials.AZURE_OPENAI_ENDPOINT
```

```python
from openai import AzureOpenAI

deployment_name='gpt-4o' #This will correspond to the custom name you chose for your deployment when you deployed a model.
deployment_name2='gpt-4o-mini' #This will correspond to the custom name you chose for your deployment when you deployed a model. 
```

```python
client = AzureOpenAI(
    api_key=os.getenv("AZURE_OPENAI_API_KEY"),  
    api_version="2024-02-15-preview",
    azure_endpoint = os.getenv("AZURE_OPENAI_ENDPOINT")
    )
```

# ChatGPT

```python
start_phrase = 'What is your name? '

response = client.chat.completions.create(
    messages=[
            {"role": "system", "content": "Your name is Jürg and you are very, very polite assistant and you talk like a pirate."},
            {"role": "user", "content": start_phrase}
    ],
    model = deployment_name,
    #max_tokens = 20
)

response
print(f"Q: {start_phrase}\nA: {response.choices[0].message.content}")
```

```python
start_phrase = 'Where do koala bears live? '

response = client.chat.completions.create(
    messages=[
            {"role": "system", "content": "Your name is Jürg and you are very, very polite assistant and you talk like a pirate."},
            {"role": "user", "content": start_phrase}
    ],
    model = deployment_name2,
    #max_tokens = 20
)

response
print(f"Q: {start_phrase}\nA: {response.choices[0].message.content}")
```

# Embeddings

```python
def generate_embeddings(text, model="text-embedding-3-large"): # model = "deployment_name"
    return client.embeddings.create(input = [text], model=model).data[0].embedding
```

```python
emb = generate_embeddings("What is my embedding?")
len(emb), emb
```

```python
emb2 = generate_embeddings("What is my embedding?", model="text-embedding-ada-002")
len(emb2), emb2
```
